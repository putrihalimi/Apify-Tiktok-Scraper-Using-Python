````python
# Feature engineering lebih proper
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Pastikan datetime
df["createTimeISO"] = pd.to_datetime(df["createTimeISO"])

# Engagement
df["engagement"] = (
    df["diggCount"] +
    df["commentCount"] +
    df["shareCount"] +
    df["collectCount"]
)

# Engagement rate (hindari division by zero)
df = df[df["playCount"] > 0]
df["engagement_rate"] = df["engagement"] / df["playCount"]

# Extract time info
df["hour"] = df["createTimeISO"].dt.hour
df["day"] = df["createTimeISO"].dt.day_name()
df["is_weekend"] = df["day"].isin(["Saturday","Sunday"])
````

````python
# üìä 3Ô∏è‚É£ Weekend vs Weekday (Dengan Uji Statistik)
# Pisahkan datanya:
weekend = df[df["is_weekend"] == True]["engagement_rate"]
weekday = df[df["is_weekend"] == False]["engagement_rate"]

t_stat, p_value = stats.ttest_ind(weekend, weekday, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
````

````python
# Pastikan data siap
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Datetime
df["createTimeISO"] = pd.to_datetime(df["createTimeISO"])

# Engagement
df["engagement"] = (
    df["diggCount"] +
    df["commentCount"] +
    df["shareCount"] +
    df["collectCount"]
)

df = df[df["playCount"] > 0]
df["engagement_rate"] = df["engagement"] / df["playCount"]

# Time features
df["hour"] = df["createTimeISO"].dt.hour

# Music binary
df["music_original_binary"] = df["musicMeta.musicOriginal"].astype(int)
````

````python
# üî• 1Ô∏è‚É£ Apakah Jam Posting Berpengaruh?: Gunakan ANOVA sederhana (karena hour kategorikal)
import statsmodels.formula.api as smf

model_hour = smf.ols("engagement_rate ~ C(hour)", data=df).fit()
anova_hour = sm.stats.anova_lm(model_hour, typ=2)

print(anova_hour)
# üëâ Kalau p-value < 0.05 ‚Üí jam signifikan.
````

````python
# üî• 2Ô∏è‚É£ Apakah Durasi Video Berpengaruh?: Regresi linear sederhana:
model_duration = smf.ols(
    "engagement_rate ~ Q('videoMeta.duration')",
    data=df
).fit()

print(model_duration.summary())
````

Lihat:

p-value durasi

sign coefficient (+ atau -)

````python
# üî• 3Ô∏è‚É£ Apakah Musik Original Berpengaruh?
model_music = smf.ols(
    "engagement_rate ~ music_original_binary",
    data=df
).fit()

print(model_music.summary())
````

Kalau:

p < 0.05 ‚Üí musik berpengaruh signifikan

coefficient positif ‚Üí original lebih tinggi

coefficient negatif ‚Üí non-original lebih tinggi

````python
# üî• 4Ô∏è‚É£ Model Gabungan (Paling Powerful)
# Sekarang kita uji semuanya sekaligus
model_full = smf.ols(
    "engagement_rate ~ C(hour) + Q('videoMeta.duration') + music_original_binary",
    data=df
).fit()

print(model_full.summary())
````

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df["performance_cluster"] = kmeans.fit_predict(scaled_data)
```
